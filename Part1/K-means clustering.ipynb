```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.cluster import KMeans

%matplotlib notebook
```

```python
data, _= make_classification(n_samples=200,
                                    n_features=2,
                                    n_informative=2,
                                    n_redundant=0,
                                    random_state=0,
                                    n_clusters_per_class=1,
                                    class_sep = 0.8)
plt.figure()
plt.scatter  (data[:,0], data[:,1])
```

```python
J = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=200).fit(data)
    J[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(J.keys()), list(J.values()))
plt.xlabel("Number of clusters")
plt.ylabel("Cost function")
plt.show()
```

```python
clf = KMeans(n_clusters=3)
clf.fit(data)
predicted= clf.predict(data)
x, y = np.array(list((zip(*data))))
```

```python
colors = np.array(['green','orange','grey', 'brown'])
plt.figure()
# draw samples 
plt.scatter(x, y, c= colors[predicted])
# draw centroids 
for i, c in enumerate (clf.cluster_centers_):    
    plt.plot(c[0], c[1], marker = 'x', color= colors[i], markersize=14) 
    
plt.title ('K-means')
```
